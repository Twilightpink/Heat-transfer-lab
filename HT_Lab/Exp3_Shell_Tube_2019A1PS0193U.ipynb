### Theory:
A transfer type of heat exchanger is one in which both fluids pass simultaneously through the device and heat is transferred through separating walls. In practice, most of the heat exchangers used are transfer type ones.
The transfer type exchangers are further classified according to flow arrangement as -

1. Single-Single Pass
2. 1 - 2 Parallel-Counter Flow Exchanger
3. 2 - 2 Pass

A simple example of transfer type of heat exchanger can be in the form of a tube type arrangement in which one of the fluids is flowing through the inner tube and the other through the annulus surrounding it. The heat transfer takes place across the walls of the inner tube.

### Formulae:


1. $$q_{av} = \frac{q_{h}+q_{c}}{2}$$

where $q_{h} = m_{h}C_{ph}(T_{hi}-T_{ho})$ and $q_{c} = m_{c}C_{pc}(T_{ci}-T_{co})$


2. Heat Transfer Area $$ A_{o} = N*\pi *d_{o}*L$$

3. LMTD $$\Delta T_{m}=\frac{\Delta T_{i}-\Delta T_{o}}{\ln \left(\Delta T_{i} 
/ \Delta T_{o}\right)}$$
$$\Delta T_{i}=T_{h i}-T_{c o}$$  $$\quad \Delta T_{o}=T_{h o}-T_{c i}$$

4. Overall Heat Transfer Co-efficient
$$U_{o} = \frac{q_{av}}{F A_{o}\Delta T_{m}}$$


5. Effectiveness of the heat exchanger
$$\epsilon = \frac{q_{av}}{q_{max}} = \frac{q_{av}}{(mCp)_{min}(T_{hi}-T_{ci})}$$

### Observations:

```python
#Experiment
N = 24           #number of tubes
do = 208e-3       #inner diameter
L = 500e-3       #length of tube
#HOT water side
W_h = np.array([200,200,200,200])
Thi = np.array([48.1,48,47.9,48.2])
Tho = np.array([43.3,42.4,43.5,43.2])
W_c = np.array([50,75,100,125])
Tci = np.array([29.1,29.9,35.3,34.9])
Tco = np.array([43.2,41.1,41.8,41.2])
F = np.array([0.8154,0.8656,0.8964,0.9021])         #From Che calc

obs = pd.DataFrame({'Hot Flow rate (LPH)':W_h,
                   'Th_i':Thi,
                   'Th_o':Tho,
                   'Cold Flow rate (LPH)':W_c,
                   'Tc_i':Tci,
                   'Tc_o':Tco})
print(obs)
```

```python
Tc = (Tci+Tco)/2
Th = (Thi+Tho)/2
cp_h = 4.174       #J/kg K
cp_c = 4.174       #J/kg K
rho_h = 989.83     #kg/m3
rho_c = 994
mh = W_h*rho_h/(1000*3600)
mc = W_c*rho_c/(1000*3600)
qc = mc*cp_c*(Tco-Tci)
qh = mh*cp_h*(Thi-Tho)
qav = (qc + qh)/2
Ao = N*np.pi*do*L
delTi = Thi - Tco
delTo = Tho - Tci
delTm = (delTi - delTo)/np.log(delTi/delTo)
Uo = qav/(Ao*delTm*F)
qmax = mc
epsilon = qav/(mc*cp_c*(Thi-Tci))                   #(mc)min is mc*cp_c
print(epsilon)
```

### Calculations:

```python
result = pd.DataFrame({'q_H':qh,
                      'q_C':qc,
                      'q_av':qav,
                      'delTm':delTm,
                       'F':F,
                      'Uo':Uo,
                      'epsilon':epsilon})
print(result)
```

### Graph:

```python
plt.xlabel('Cold water flow rate (LPH)',fontsize=15)
plt.ylabel('Overall heat transfer coefficient U_o, effectiveness \u03B5',fontsize=15)
plt.rcParams["figure.figsize"] = (12,9)
plt.plot(W_c,Uo,label='Uo',color='red')
plt.plot(W_c,epsilon,label='\u03B5',color='blue')
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.grid()
plt.legend(loc="best")
plt.show()
```

### Industrial Applications:

Shell and tube heat exchangers are the most widely used heat exchanger type in a number of different industries. These heat exchangers are constructed using several parallel heat exchanger tubes. Tube sheets are used to hold these heat exchanger tubes firmly in place. These tubes are placed in a shell. The pattern of flow through a shell and tube heat exchanger can have a number of different configuration options. Be it any configuration, the design of these heat exchangers is such that one fluid passes through the shell (and is called shell side fluid), while other fluid passes through the tubes (and is called the tube side fluid). The heat exchangers have tube walls, which allow the exchange of heat between two fluids.
